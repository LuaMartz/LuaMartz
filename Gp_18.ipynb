**Operadores**

♦ Operadores aritméticos

```python
#SUMA
x = 10
y = 8
resul = x + y
print (x, "+", y, "=", resul)

#RESTA
resul = x - y
print (x, "-", y, "=", resul)

#DIVISION
#2 valores
resul = x / y
print (x, "/", y, "=", resul)
#2 valores enteros
resul = x // y
print (x, "/", y, "=", resul)

#MULTIPLICACION
resul = x * y
print (x, "x", y, "=", resul)

#RESTO/MODULO/RESIDUO
resul = x % y
print (x, "%", y, "=", resul)

#EXPONENCIACION
resul = x ** y
print (x, "^", y)
```

♦ Asignación

```python
x = 15
y = 10

#Asignar con suma +=
x += y      #x = x + y
print ((x-y), "+", y, "=", x )

#Asignar con resta -=
x -= y      #x = x - y
print (x, "-", y, "=", (x-y))

#Asignar con multiplicacion *=
x *= y      #x = x * y
print ((x//y), "x", y, "=", x )

#Asignar con division /=
x //= y      #x = x / y
print (x, "-", y, "/", (x/y))

#Asignar con residuo %=
x %= y      #x = x % y
print ((x+y), "%", y, "=", x )
```

♦ Lógicos: Se utilizan para combinar sentencias condicionales

```python
#and (y)
q = 5
print ("Y")
print (q > 5 and q < 0)

#or (o)
print ("O")
print (q > 5 or q < 10)

#not (no)
print ("NOT")
print (not (q > 2 and q <7))
```

♦ Relacionales: Se utilizan para compara dos valores

```python
m = 45
n = 23

#Igualdad
print ("Igual que:")
print(" ", m==n)

#Mayor que >
print ("Mayor que:")
print (" ", m > n)

#Menor que <
print ("Menor que:")
print (" ", m < n)

#Mayor igual que >=
print ("Mayor igual que:")
print (" ", m >= n)

#Menor igual que <=
print ("Menor igual que:")
print (" ", m <= n)

#Diferente que !=
print ("Diferente que")
print (" ", m != n)
```
