**Tipos de datos** ❤


♦ Numeros Enteros: Positivos o negativos, sin decimales (int)

```python
#Entradas
num1 : int = 27
num2 : int = 50
#Salida
suma : int = num1 + num2
#Proceso
print ("Ejemplo:", num1, "+", num2, "=", suma)
```

♦ Numeros Reales: Positivos o negativos, con decimales (float)

```python

peso : float = 55.6
print ("Ejemplo: El peso es:", peso)
print (nombre1, nombre2, apellido1, apellido2, "pesa:", peso, "kg")
```

♦ Booleanos: Toma dos valores True o False (bool)

```python
esColombiano : bool = True
print ("Ejemplo: Es Colombiana??", esColombiano)
```

**Operadores**

♦ Operadores aritméticos

```python
#SUMA
x = 10
y = 8
resul = x + y
print (x, "+", y, "=", resul)

#RESTA
resul = x - y
print (x, "-", y, "=", resul)

#DIVISION
#2 valores
resul = x / y
print (x, "/", y, "=", resul)
#2 valores enteros
resul = x // y
print (x, "/", y, "=", resul)

#MULTIPLICACION
resul = x * y
print (x, "x", y, "=", resul)

#RESTO/MODULO/RESIDUO
resul = x % y
print (x, "%", y, "=", resul)
```
